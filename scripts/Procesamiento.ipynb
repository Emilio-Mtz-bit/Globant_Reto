```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/data_globant.csv")
```

```python
data["Engagement"].value_counts()
```

```python
data.duplicated().sum()
data['Date'] = data['Date'].str.lstrip("0")
(data['Date'].unique())
##print(len(data['Team Name'].unique()))
```

```python
data["Date_timestamp"] = pd.to_datetime(data['Date'], format="%d%b%y")
```

```python
data.info()
```

```python
data[data["Name"]=="Alberto Bravo"]
```

```python
data[data[['Date','Name']].duplicated()]
data[['Date','Name']].duplicated().sum()
```

```python
mean_engagement = data.groupby(["Date_timestamp","Team Name"])["Engagement"].mean().reset_index()
mean_engagement
```

```python
# 1. Definir las funciones de agregaciÃ³n
# 'mean' para 'engagement' y 'first' para el resto (asumiendo 'Date' y 'Name' son las claves)
agg_functions = {
    col: 'first' for col in data.columns if col not in ['Date', 'Name', 'Engagement']
}
agg_functions['Engagement'] = 'mean'

# 2. Agrupar, agregar y resetear el Ã­ndice para que 'Date' y 'Name' vuelvan a ser columnas
data_final = data.groupby(['Date', 'Name'], as_index=False).agg(agg_functions)
```

```python
data_final[['Date','Name']].duplicated().sum()
data_final[(data_final['Date']=="16May23") & (data_final["Name"]=="MarÃ­a Ceballos")]
data_final.sort_valueshead()
```

```python
data_final.to_csv("../data/Final_Data.csv",index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ajusta el tamaÃ±o del grÃ¡fico
plt.figure(figsize=(12, 6))

for i in mean_engagement['Team Name'].unique():

    # Crea el grÃ¡fico de lÃ­neas
    sns.lineplot(
        data=mean_engagement.sort_values('Date_timestamp')[mean_engagement['Team Name']==i],
        x='Date_timestamp',
        y='Engagement',
        hue='Team Name'  # Esto crea una lÃ­nea por cada equipo
    )

    # Mueve la leyenda fuera del grÃ¡fico
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

    # AÃ±ade un tÃ­tulo
    plt.title(f'Mean Engagement Over Time of {i}')
    plt.show()
```
